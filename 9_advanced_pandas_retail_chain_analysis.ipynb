### part one | datetime operations

#### 1. Converting and Working with Dates:

```python
# Load all three CSV files and convert date columns to proper DateTime format using pd.to_datetime()

df_customers['registration_date'] = pd.to_datetime(df_customers['registration_date'])

df_products['launch_date'] = pd.to_datetime(df_products['launch_date'])

df_sales['transaction_date'] = pd.to_datetime(df_sales['transaction_date'])
```

```python
# Extract useful components from the sales transaction dates: year, month, and day of week

df_customers['registration_year'] = df_customers['registration_date'].dt.year  # noqa
df_customers['registration_month'] = df_customers['registration_date'].dt.month  # noqa
df_customers['registration_day'] = df_customers['registration_date'].dt.day_name()  # noqa

df_products['launch_year'] = df_products['launch_date'].dt.year  # noqa
df_products['launch_month'] = df_products['launch_date'].dt.month  # noqa
df_products['launch_day'] = df_products['launch_date'].dt.day_name()  # noqa

df_sales['transaction_year'] = df_sales['transaction_date'].dt.year  # noqa
df_sales['transaction_month'] = df_sales['transaction_date'].dt.month  # noqa
df_sales['transaction_day'] = df_sales['transaction_date'].dt.day_name()  # noqa

df_sales.head()
```

```python
# Display the date range covered by your sales data (earliest and latest dates)

print(df_sales['transaction_date'].min())
print(df_sales['transaction_date'].max())
```

#### 2. Basic Time Analysis:

```python
# Find the busiest day of the week for sales (by counting transactions)

busiest_day = df_sales.groupby('transaction_day')['sale_id'].sum().sort_values(ascending=False)
print(busiest_day)

# the busiest day is Friday
```

```python
# Determine which month had the most transactions

busiest_month = df_sales.groupby('transaction_month')['sale_id'].sum().sort_values(ascending=False)
print(busiest_month)

# the busiest month is December (12)
```

```python
# Count how many sales happened in each year

busiest_year = df_sales.groupby('transaction_year')['sale_id'].sum().sort_values(ascending=False)
print(busiest_year)

# the busiest year is 2023
```

### part two | groupby and aggregation

#### 1. Product Analysis:

```python
# Group products by category and calculate the average price for each category

average_price = df_products.groupby('category')['price'].mean()
print(average_price)
```

```python
# Count how many products are in each category

number_of_products = df_products.groupby('category')['product_id'].sum()
print(number_of_products)
```

#### 2. Customer Analysis:

```python
# Group customers by state and find the average age in each state

avg_age_by_state = df_customers.groupby('state')['age'].mean()
print(avg_age_by_state)
```

```python
# Calculate total spending per customer (sum up all their purchases)

spend_per_customer = df_sales.groupby('customer_id')['total_amount'].sum()
print(spend_per_customer)
```

```python
# Find customers who have made the most purchases

customer_purchases = df_sales.groupby('customer_id')['sale_id'].sum().sort_values(ascending=False).head(5)
print(customer_purchases)
```

#### 3. Sales Analysis:

```python
# Calculate total sales revenue by month

monthly_revenue = df_sales.groupby('transaction_month')['total_amount'].sum()
print(monthly_revenue)
```

```python
# Find the average transaction amount by day of the week

avg_amount_day_of_week = df_sales.groupby('transaction_day')['total_amount'].mean().sort_values()
print(avg_amount_day_of_week)
```

```python
# Determine which product category generates the most total revenue

sales_per_category = df_products.groupby('category')['price'].sum().sort_values(ascending=False).head(1)
print(sales_per_category)
```

### part three | basic data insights

#### 1. Simple Comparisons:

```python
# Find which state has the most customers

state_most_customers = df_customers.groupby('state')['customer_id'].sum().sort_values(ascending=False).head(1)
print(state_most_customers)
```

```python
# Identify the top 5 customers by total spending

top_spending_customers = df_sales.groupby('customer_id')['total_amount'].sum().sort_values(ascending=False).head(5)
print(top_spending_customers)
```

```python
# Determine which day of the week has the highest average transaction value

highest_average_grossing_day = df_sales.groupby('transaction_day')['total_amount'].mean().sort_values(ascending=False).head(1)
print(highest_average_grossing_day)
```
