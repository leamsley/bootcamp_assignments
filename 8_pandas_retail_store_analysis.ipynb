## pt. 2 | data cleaning

### handle missing values

```python
# identify all missing values in each dataset
df_customers.isnull()
```

```python
df_products.isnull()
```

```python
df_stores.isnull()
```

```python
df_sales.isnull()
```

```python
df_inventory.isnull()
```

```python
type(df_products['weight'])
```

```python
# for numerical columns with missing values, replace them with the column mean

df_products['weight'] = df_products['weight'].fillna(df_products['weight'].mean())
df_stores['size_sqft'] = df_stores['size_sqft'].fillna(df_stores['size_sqft'].mean())
df_inventory['reorder_level'] = df_inventory['reorder_level'].fillna(df_inventory['reorder_level'].mean())
```

```python
# for categorical columns with missing values, replace them with the most frequent value

# Categorical columns with missing data:
# df_customers | zip_code
# df_sales | payment_method

df_customers['zip_code'] = df_customers['zip_code'].fillna(df_customers['zip_code'].mode()[0])
df_sales['payment_method'] = df_sales['payment_method'].fillna(df_sales['payment_method'].mode()[0])
```

```python
# for date columns with missing values, use forward fill or backward fill as appropriate

df_customers['registration_date'] = df_customers['registration_date'].bfill()
df_stores['opening_date'] = df_stores['opening_date'].ffill()
df_sales['date'] = df_sales['date'].bfill()
df_inventory['last_restock_date'] = df_inventory['last_restock_date'].ffill()
```

### removing duplicates

```python
# check for and remove any duplicate entries in the customers and products dataframes

# for df_customers: check for duplicates in the first_name and last_name columns
df_customers.duplicated(subset=['first_name', 'last_name'])
```

```python
# betty lewis appears twice, so we will keep the first instance and remove the other entry
df_customers = df_customers.drop_duplicates(subset=['first_name'])
```

```python
# for df_products: check for duplicates in the product_id and product_name columns
df_products.duplicated(subset=['product_id', 'product_name'])
# there are no duplicates in df_products
```
